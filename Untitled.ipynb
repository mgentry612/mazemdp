```python
from random import shuffle, randrange
from mdpstate import MDPState
from copy import deepcopy
import random
from collections import deque
```

```python
random.seed(1)
```

```python
def printmaze(maze):
    for row in maze:
        print(''.join(row))

def get_s():
    s = ""
    for (a, b) in zip(hor, ver):
        s += ''.join(a + ['\n'] + b + ['\n'])

    s = s.split('\n')
    s = [list(row) for row in s]
    del s[len(s)-1]
    del s[len(s)-1]
    
    return s
```

```python
w = 8
h = 6

visited = [[0] * w + [1] for _ in range(h)] + [[1] * (w + 1)]
ver = [["# "] * w + ['#'] for _ in range(h)] + [[]]
hor = [["##"] * w + ['#'] for _ in range(h + 1)]

to_visit = deque([(randrange(h), randrange(w))])

while len(to_visit) > 0:

    current = to_visit.pop()
    y = current[0]
    x = current[1]
    
    visited[y][x] = 1

    neighbors = [(x - 1, y), (x, y + 1), (x + 1, y), (x, y - 1)]
    
    shuffle(neighbors)
    
    for (xx, yy) in neighbors:
        if visited[yy][xx]: continue
        to_visit.append((yy, xx))
    
    if len(to_visit) > 1:
        
        current_neighbor = to_visit[-1]

        yy = current_neighbor[0]
        xx = current_neighbor[1]

        if xx == x: hor[max(y, yy)][x] = "# "
        if yy == y: ver[y][max(x, xx)] = "  "
            
        printmaze(get_s())
        print("\n")

printmaze(get_s())
```
